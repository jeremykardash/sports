```python
import os
import pandas as pd
```

```python
file = os.path.join("resources/2019.csv")
```

```python
data = pd.read_csv(file)
```

```python
data
```

```python
# reading csv file from url  
# dropping null value columns to avoid errors 
data.dropna(inplace = True) 
  
# new data frame with split value columns 
new = data["Player"].str.split("\\", n = 1, expand = True)  
  
# making separate first name column from new data frame 
data["Name"]= new[0] 
  
# df display 
data
```

```python
#Drop duplicates for traded players
clean_stats = data.drop_duplicates(subset=['Name'], keep='first')
#clean_stats
```

```python
#Drop players under 500 minutes played
clean_stats = clean_stats.loc[clean_stats['MP'] > 499]
#clean_stats
```

```python
#Column check for cleanup
#list(clean_stats)
```

```python
#Clean for columns
df = clean_stats[['Name', 'Pos', 'Age', 'Tm', 'PTS', 'TRB', 'AST', '3P', 'FG%', 'FT%', 'STL', 'BLK', 'TOV']]

#Clean for position
guards = df.loc[(df['Pos'] == 'PG') | (df['Pos'] == 'SG')]
forwards = df.loc[(df['Pos'] == 'SF') | (df['Pos'] == 'PF')]
centers = df.loc[df['Pos'] == 'C']
```

```python
#df['zscore'] = (df.a - df.a.mean())/df.a.std(ddof=0)
guards['PTS Z'] = (guards['PTS'] - guards['PTS'].mean())/guards['PTS'].std(ddof=0)
guards['TRB Z'] = (guards['TRB'] - guards['TRB'].mean())/guards['TRB'].std(ddof=0)
guards['AST Z'] = (guards['AST'] - guards['AST'].mean())/guards['AST'].std(ddof=0)
guards['3P Z'] = (guards['3P'] - guards['3P'].mean())/guards['3P'].std(ddof=0)
guards['FG% Z'] = (guards['FG%'] - guards['FG%'].mean())/guards['FG%'].std(ddof=0)
guards['FT% Z'] = (guards['FT%'] - guards['FT%'].mean())/guards['FT%'].std(ddof=0)
guards['STL Z'] = (guards['STL'] - guards['STL'].mean())/guards['STL'].std(ddof=0)
guards['BLK Z'] = (guards['BLK'] - guards['BLK'].mean())/guards['BLK'].std(ddof=0)
guards['TOV Z'] = (guards['TOV'] - guards['TOV'].mean())/guards['TOV'].std(ddof=0)
guards
```

```python
forwards['PTS Z'] = (forwards['PTS'] - forwards['PTS'].mean())/forwards['PTS'].std(ddof=0)
forwards['TRB Z'] = (forwards['TRB'] - forwards['TRB'].mean())/forwards['TRB'].std(ddof=0)
forwards['AST Z'] = (forwards['AST'] - forwards['AST'].mean())/forwards['AST'].std(ddof=0)
forwards['3P Z'] = (forwards['3P'] - forwards['3P'].mean())/forwards['3P'].std(ddof=0)
forwards['FG% Z'] = (forwards['FG%'] - forwards['FG%'].mean())/forwards['FG%'].std(ddof=0)
forwards['FT% Z'] = (forwards['FT%'] - forwards['FT%'].mean())/forwards['FT%'].std(ddof=0)
forwards['STL Z'] = (forwards['STL'] - forwards['STL'].mean())/forwards['STL'].std(ddof=0)
forwards['BLK Z'] = (forwards['BLK'] - forwards['BLK'].mean())/forwards['BLK'].std(ddof=0)
forwards['TOV Z'] = (forwards['TOV'] - forwards['TOV'].mean())/forwards['TOV'].std(ddof=0)
forwards
```

```python
centers['PTS Z'] = (centers['PTS'] - centers['PTS'].mean())/centers['PTS'].std(ddof=0)
centers['TRB Z'] = (centers['TRB'] - centers['TRB'].mean())/centers['TRB'].std(ddof=0)
centers['AST Z'] = (centers['AST'] - centers['AST'].mean())/centers['AST'].std(ddof=0)
centers['3P Z'] = (centers['3P'] - centers['3P'].mean())/centers['3P'].std(ddof=0)
centers['FG% Z'] = (centers['FG%'] - centers['FG%'].mean())/centers['FG%'].std(ddof=0)
centers['FT% Z'] = (centers['FT%'] - centers['FT%'].mean())/centers['FT%'].std(ddof=0)
centers['STL Z'] = (centers['STL'] - centers['STL'].mean())/centers['STL'].std(ddof=0)
centers['BLK Z'] = (centers['BLK'] - centers['BLK'].mean())/centers['BLK'].std(ddof=0)
centers['TOV Z'] = (centers['TOV'] - centers['TOV'].mean())/centers['TOV'].std(ddof=0)
centers
```

```python
guards["Z Total"] = guards['PTS Z'] + guards['TRB Z'] + guards['AST Z'] + guards['3P Z'] + guards['FG% Z'] + guards['FT% Z'] + guards['STL Z'] + guards['BLK Z'] - guards['TOV Z']
z_score_g = guards[['Name', "Pos", 'Z Total']].sort_values(by=["Z Total"], ascending=False)
z_score_g.reset_index()
z_score_g = z_score_g[['Name', 'Pos', 'Z Total']]
z_score_g.head(20)
```

```python
forwards["Z Total"] = forwards['PTS Z'] + forwards['TRB Z'] + forwards['AST Z'] + forwards['3P Z'] + forwards['FG% Z'] + forwards['FT% Z'] + forwards['STL Z'] + forwards['BLK Z'] - forwards['TOV Z']
z_score_f = forwards[['Name', "Pos", 'Z Total']].sort_values(by=["Z Total"], ascending=False)
z_score_f.reset_index()
z_score_f = z_score_f[['Name', 'Pos', 'Z Total']]
z_score_f.head(20)
```

```python
centers["Z Total"] = centers['PTS Z'] + centers['TRB Z'] + centers['AST Z'] + centers['3P Z'] + centers['FG% Z'] + centers['FT% Z'] + centers['STL Z'] + centers['BLK Z'] - centers['TOV Z']
z_score_c = centers[['Name', "Pos", 'Z Total']].sort_values(by=["Z Total"], ascending=False)
z_score_c.reset_index()
z_score_c = z_score_c[['Name', 'Pos', 'Z Total']]
z_score_c.head(20)
```

```python
z_score_g.to_csv("output/guards_2019")
```
