```python
import os
import pandas as pd
```

```python
file = os.path.join("output/ranking_2019_per.csv")
```

```python
z_score = pd.read_csv(file)
#z_score.head(30)
```

```python
file_2 = os.path.join("resources/fantasy_rankings.csv")
```

```python
fantasy = pd.read_csv(file_2)
#fantasy
```

```python
z_score["Name"] = z_score["Name"].astype('string')
```

```python
fantasy["Player"] = fantasy["Player"].astype('string')
```

```python
from unidecode import unidecode
```

```python
z_score["Name"] = z_score["Name"].apply(unidecode)
```

```python
try:
    fantasy["Player"] = fantasy["Player"].apply(unidecode)
except:
    pass
```

```python
fantasy = fantasy[["Rank", "Player"]]
```

```python
merged = pd.merge(fantasy, z_score, left_on="Player", right_on="Name", how="inner")
merged
```

```python
rename = merged.rename(columns={
    'Rank_x': "Fantasy Ranking",
    'Rank_y': "Z-Score Ranking",
})
rename
```

```python
final_df = rename[["Player", "Pos", "Fantasy Ranking", "Z-Score Ranking", 'Z Total']]
final_df['Difference'] = final_df["Fantasy Ranking"] - final_df["Z-Score Ranking"]
final_df.to_csv("output/delta_fantasy_2019_per.csv", index=False)
```
