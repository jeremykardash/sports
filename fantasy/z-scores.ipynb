```python
import os
import pandas as pd
```

```python
guard_19 = os.path.join("output/guards_2019")
guard_18 = os.path.join("output/guards_2018")
guard_17 = os.path.join("output/guards_2017")
forwards_19 = os.path.join('output/forwards_2019')
forwards_18 = os.path.join('output/forwards_2018')
forwards_17 = os.path.join('output/forwards_2017')
centers_19 = os.path.join('output/centers_2019')
centers_18 = os.path.join('output/center_2018')
centers_17 = os.path.join('output/centers_2017')
```

```python
df_g_19 = pd.read_csv(guard_19)
df_g_18 = pd.read_csv(guard_18)
df_g_17 = pd.read_csv(guard_17)
df_f_19 = pd.read_csv(forwards_19)
df_f_18 = pd.read_csv(forwards_18)
df_f_17 = pd.read_csv(forwards_17)
df_c_19 = pd.read_csv(centers_19)
df_c_18 = pd.read_csv(centers_18)
df_c_17 = pd.read_csv(centers_17)
```

```python
guard_merge = pd.merge(df_g_19, df_g_18, on="Name")
```

```python
forward_merge = pd.merge(df_f_19, df_f_18, on="Name")
```

```python
center_merge = pd.merge(df_c_19, df_c_18, on="Name")
```

```python
guard = guard_merge[["Name", "Pos_x", "Z Total_x", "Z Total_y"]]
```

```python
forward = forward_merge[["Name", "Pos_x", "Z Total_x", "Z Total_y"]]
```

```python
center = center_merge[["Name", "Pos_x", "Z Total_x", "Z Total_y"]]
```

```python
guard = guard.rename(columns={
    "Pos_x": "Pos",
    "Z Total_x": "Z Score 2019",
    "Z Total_y": "Z Score 2018"
})
guard["Z Score Difference"] = guard["Z Score 2019"] - guard["Z Score 2018"]
guard = guard.sort_values(by=["Z Score Difference"], ascending=False)
```

```python
forward = forward.rename(columns={
    "Pos_x": "Pos",
    "Z Total_x": "Z Score 2019",
    "Z Total_y": "Z Score 2018"
})
forward["Z Score Difference"] = forward["Z Score 2019"] - forward["Z Score 2018"]
forward = forward.sort_values(by=["Z Score Difference"], ascending=False)
```

```python
center = center.rename(columns={
    "Pos_x": "Pos",
    "Z Total_x": "Z Score 2019",
    "Z Total_y": "Z Score 2018"
})
center["Z Score Difference"] = center["Z Score 2019"] - center["Z Score 2018"]
center = center.sort_values(by=["Z Score Difference"], ascending=False)
center
```
